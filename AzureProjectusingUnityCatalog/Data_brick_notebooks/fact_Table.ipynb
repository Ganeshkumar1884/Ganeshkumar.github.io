```python
from pyspark.sql.types import *
from pyspark.sql.functions import *
df_src_fact=spark.sql('''select branch_id, dealer_id, model_id,date_id, revenue, units_sold, date_id,average_sales  from parquet.`abfss://silver@adlsdevgk.dfs.core.windows.net/`
                  ''')
df_src_fact.display()





              
```

```python
df_dim_model=spark.sql(''' select * from delta.`abfss://gold@adlsdevgk.dfs.core.windows.net/dim_model`
                       ''')
df_dim_date=spark.sql(''' select * from delta.`abfss://gold@adlsdevgk.dfs.core.windows.net/dim_date`
                       ''')
df_dim_branch=spark.sql(''' select * from delta.`abfss://gold@adlsdevgk.dfs.core.windows.net/dim_branch`
                       ''')

df_dim_dealer=spark.sql(''' select * from delta.`abfss://gold@adlsdevgk.dfs.core.windows.net/dim_dealer`
                       ''')
```

```python
df_dim_model.display()
```

```python
df_fact_final1=df_src_fact.join(df_dim_model,df_src_fact.model_id==df_dim_model.model_id,'left')\
                        .join(df_dim_branch,df_dim_branch.branch_id==df_src_fact.branch_id,'left')\
                            .join(df_dim_dealer,df_dim_dealer.dealer_id==df_src_fact.dealer_id,'left')\
                                .join(df_dim_date,df_dim_date.date_id==df_src_fact.date_id,'left')\
                            .select(df_src_fact.revenue, df_src_fact.units_sold, df_src_fact.average_sales,df_dim_model.dim_model_key,df_dim_branch.branch_key,df_dim_dealer.dealer_key,df_dim_date.date_key)
```

## Fact table upsert logic 

```python
from delta.tables import DeltaTable
if spark.catalog.tableExists('car_sales_data.gold.fact'):
    delta_tbl=DeltaTable.forPath(spark,'abfss://gold@adlsdevgk.dfs.core.windows.net/fact')
    delta_tbl.alias('tgt').merge(df_fact_final1.alias('src'), "tgt.dim_model_key=src.dim_model_key and tgt.branch_key=src.branch_key and tgt.dealer_key=src.dealer_key and tgt.date_key=src.date_key").whenMatchedUpdateAll()\
    .whenNotMatchedInsertAll()\
    .execute()
else :
    df_fact_final1.write.format('delta')\
        .option('path', 'abfss://gold@adlsdevgk.dfs.core.windows.net/fact')\
        .mode('overwrite')\
        .saveAsTable('car_sales_data.gold.fact')
```
